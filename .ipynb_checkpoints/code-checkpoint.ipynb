As we can see ,our dataset contains missing values, particularly in the 'Experience level' and 'Salary' columns. The code cell below  to calculate the percentage of missing values in each column.
Since the count of missing values is small (1 for each column), we may consider removing the rows with missing values using.

```python
# Drop rows with missing values in specified columns
df = df.dropna(subset=['Company', 'Job Title', 'Location', 'Job Type'])
df.isnull().sum()
```

Now, let's address the 'Experience level' column. Our suggestion is to initially fill in 'Experience level' values for rows where we have a valid 'Salary'.
Subsequently, we can infer the 'Experience level' based on the salary values, completing the missing labels for the remaining rows.

```python
# Convert 'Salary' to numeric format
df['Salary'] = pd.to_numeric(df['Salary'].str.replace('[^\d.]', '', regex=True), errors='coerce')

# Define the number of quantiles or bins for salary ranges
num_quantiles = 4  # as we have 4 experience levels

# Create salary ranges using qcut
df['Salary Range'] = pd.qcut(df['Salary'], q=num_quantiles, labels=False)

# Map quantile labels to experience levels
experience_levels = ['Entry-level', 'Mid-level', 'Senior-level', 'Executive-level']
# Iterate over rows
for index, row in df.iterrows():
    # Check if 'Experience level' is null
    if pd.isnull(row['Experience level']):
        # Check if 'Salary' is not null
        if not pd.isnull(row['Salary']):
            # Map 'Salary' to 'Experience level'
            df.at[index, 'Experience level'] = experience_levels[int(row['Salary Range'])]
# Drop the temporary 'Salary Range' column
df = df.drop(columns=['Salary Range'])

# Display the updated DataFrame
print(df[['Experience level']])


```

```python
df.isnull().sum()
```

Now, let's handle the 'Salary' column. We propose a similar logic, calculating the mean salary for each 'Experience level'.
When encountering a row with a valid 'Experience level' but a missing 'Salary', we fill the 'Salary' with the mean of that specific 'Experience level'.


```python
# Iterate over rows
for index, row in df.iterrows():
    # Check if 'Salary' is null
    if pd.isnull(row['Salary']):
        # Check if 'Experience level' is not null
        if not pd.isnull(row['Experience level']):
            # Calculate the mean salary for the 'Experience level' range
            mean_salary = df[df['Experience level'] == row['Experience level']]['Salary'].mean()
            # Fill the null 'Salary' with the calculated mean salary
            df.at[index, 'Salary'] = mean_salary

# Display the updated DataFrame
print(df[['Experience level', 'Salary']])
```

We will proceed with trimming outliers using the Percentile method to define both the upper and lower thresholds.


```python
# lower_threshold = df['Salary'].quantile(0.05)
# upper_threshold = df['Salary'].quantile(0.95)

Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1
lower_threshold = Q1 - 1.5 * IQR
upper_threshold = Q3 + 1.5 * IQR
```

```python
df = df[(df['Salary'] >= lower_threshold) & (df['Salary'] <= upper_threshold)]
```

```python

# Plot the salary distribution by Experience level
plt.figure(figsize=(15, 8))
sns.boxplot(x='Experience level', y='Salary', data=df)
plt.xticks(rotation=45, ha='right')
plt.title('Salary Distribution by Experience level')
plt.xlabel('Job Title')
plt.ylabel('Experience level')
plt.show()
```

```python
import pycountry

# Assuming your dataset is named 'df' and the column is 'Location'
def extract_country_code(address):
    for country in pycountry.countries:
        if country.name in str(address):
            return country.alpha_2
    return None

# Apply the function to create a new 'Country Code' column
df['Country Code'] = df['Location'].apply(extract_country_code)

# Display the result
print(df[['Location', 'Country Code']])
```

### 1.  Explore Requirment of the companies

as we see we have 2600 unique requirement of the company ,we will construct a list of all the requirement of the company , because in our dataset the column Requirment of the company is a set of requirements 

```python
# Split the comma-separated values in the "Requirment of the company" column
requirements_split = df['Requirment of the company'].str.split(',')

# Create a set to store unique requirements
unique_requirements_set = set()

# Iterate through each row and update the set
for requirements_list in requirements_split:
    unique_requirements_set.update(requirements_list)

# Convert the set to a list
unique_requirements_list = list(unique_requirements_set)

#print the result 
print(unique_requirements_list)
len(unique_requirements_list)
```

this is the list of all comapany requirements for the jobs profiles, we have 286 unique requirements

```python
# Create a dictionary to store counts for each requirement
requirement_counts = {req: 0 for req in unique_requirements_list}

# Iterate over each unique requirement and count occurrences in the DataFrame
for requirement in unique_requirements_list:
    # Count occurrences of the requirement in the 'Requirment of the company' column
    count_for_requirement = df['Requirment of the company'].apply(lambda x: requirement in x.split(',')).sum()
    
    # Update the dictionary with the count
    requirement_counts[requirement] = count_for_requirement

# Find the most common requirement and its count
most_common_requirement = max(requirement_counts, key=requirement_counts.get)
count_of_most_common_requirement = requirement_counts[most_common_requirement]

# Print the most common requirement and its count
print(f"The most common requirement is '{most_common_requirement}' with {count_of_most_common_requirement} occurrences.")
```

### 2.  Explore Job Titles

```python
# Extract all job titles from the 'Company Job Title' column
all_job_titles = df['Job Title'].tolist()

# Get unique job titles using set and convert them to a list
unique_job_titles = list(set(all_job_titles))

# Print the list of unique job titles
print("List of Unique Job Titles:")
print(unique_job_titles)
```

we will search for the top 5 Job titles occured in our dataset

```python
# Count the occurrences of each job title
job_title_counts = Counter(df["Job Title"])

# Find the most common job title and its count
most_common_job_title, count = job_title_counts.most_common(1)[0]

# Print the most common job title and its count
print(f"The most common job title is '{most_common_job_title}' with {count} occurrences.")

top_n = 5

# Find the top N most common job titles and their counts
top_job_titles = job_title_counts.most_common(top_n)

# Print the top N job titles and their counts
print(f"The top {top_n} most common job titles are:")
for job_title, count in top_job_titles:
    print(f"'{job_title}' - {count} occurrences")
```
